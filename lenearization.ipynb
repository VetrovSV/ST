# Линеаризация функции случаного аргумента. Пример

```python
import pandas as pd
import numpy as np
from matplotlib.pyplot import *
import seaborn
```

```python
def bomb_range(H, c, v0):
    return v0 * (2*H / 9.81)**0.5 * (1 - 1.8e-5 * c * H)
```

```python
H = np.linspace(3500, 4500, 100)
v0 = 150
c = 1

X = bomb_range(H,c,v0)
```

```python
grid(True)
plot(H,X)
xlabel('высота H')
ylabel('относ бомбы')
```

```python
H = 4000
v0 = np.linspace( 130, 180, 100 )
c = 1

X = bomb_range(H,c,v0)

grid(True)
plot(v0,X)
xlabel('скорость')
ylabel('относ бомбы')
```

```python
H = 4000
v0 = 150
c = np.linspace(0.8, 1.2, 100)

X = bomb_range(H,c,v0)

grid(True)
plot(c,X)
xlabel('с')
ylabel('относ бомбы')
```
